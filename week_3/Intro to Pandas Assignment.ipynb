### Exercise #3: Grouping
Find the 5 players with the highest `OPS+`

```python
# Grouped dataframe by 'OPS'. and displayed the top 5 players
rs_17.groupby("OPS").max().sort_values('OPS', ascending=False).head(5)
```

```python
# Grouped dataframe by 'OPS' & 'Name'. and displayed the top 5 players
rs17_group_by_2 = rs_17.groupby(["OPS", "Name"]).max().sort_values("OPS", ascending=False)# grouping by OPS and name and sorting by  the top 5 in OPS
print(f'This DataFrame is grouped by OPS and Name {rs17_group_by_2.head(6)}')
```

```python
# Grouped dataframe by 'OPS' & 'Name'. and displayed the top 5 players
rs17_group_by_2 = rs_17.groupby(["OPS", "Name"]).max().sort_values("OPS", ascending=False)[['SLG', 'OBP']]     # grouping by OPS and name and sorting by  the top 5 in OPS
print(f'This DataFrame is grouped by OPS and Name {rs17_group_by_2.head(6)}')
```
