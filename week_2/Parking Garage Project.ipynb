# Parking Garage Project

Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP). 

Your parking gargage class should have the following methods:
- takeTicket
   - This should decrease the amount of tickets available by 1
   - This should decrease the amount of parkingSpaces available by 1
- payForParking
   - Display an input that waits for an amount from the user and store it in a variable
   - If the payment variable is not empty then ->  display a message to the user that their ticket has been paid and they have 15mins to leave
   - This should update the "currentTicket" dictionary key "paid" to True
-leaveGarage
   - If the ticket has been paid, display a message of "Thank You, have a nice day"
   - If the ticket has not been paid, display an input prompt for payment
      - Once paid, display message "Thank you, have a nice day!"
   - Update parkingSpaces list to increase by 1
   - Update tickets list to increase by 1

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

```python
class ParkingGarage:
    """
    This is a digital parking garage that will allow a driver to pay for parking, take a ticket and monitor parking capacity.
    Attributes: 
    - tickets: a list with the amount of tickets available
    - parking_spaces: a list of spaces used and/or available
    - current_ticket: a dictionary that accepts and stores tickets allowing drivers to leave the garage (leave_garage)
    - money_paid_by_driver: a variable that stores the cash amount paid for the ticket by the driver 

    Methods:
    - pay_for_parking: displays an input prompt to the driver for the amount they are going pay and outputs a message. a series of if statements guides the user 
    - take_ticket: provides a ticket to the driver upon payment if capacity has not been met. also it, += and -= tickets list and the parking_spaces list 
    - show_capacity: provides a print out of the garage data. including - max capacity, current capacity, and a list of the license plates that are currently occupying the garage
    """
    tickets = [1,1,1,1,1] # five tickets are available
    parking_spaces = [1,1,1,1,1] # Garage holds five automobiles
    license_plate_numbers = []

    
    
    
    def __init__(self,license_plate_number:str):
        """initializees Class Attribute license plate number"""
        self.license_plate_number = license_plate_number

        if len(self.parking_spaces) == 0:
            print("\nWe are at Max Capacity:\nPlease come again.")
        else:
             print(f"\nIt is your lucky day --> Auto {self.license_plate_number}! \nWe have an open spot for you!")



    def take_ticket(self):
        """Provides a ticket to the driver upon payment. Will subtract tickets from the tickets list and parking_spaces list"""
        if len(self.tickets) > 0 and len(self.parking_spaces) > 0:
            print(f"\nPlease take your ticket Auto--> {self.license_plate_number} \nThank you for parking here!!!")
            self.tickets.pop(0)
            self.parking_spaces.pop(0)
            self.license_plate_numbers.append(self.license_plate_number)
        else:
            print("\nWe are at Max Capacity:\nPlease come again.")



    def pay_for_parking(self, money_paid_by_driver:int):
            '''Accepts money from the driver to pay for parking and leave the garage. Upon payment, prompts driver of 15 minute time limit to leave'''
            self.money_paid_by_driver = money_paid_by_driver
            if self.money_paid_by_driver == 10:
                
                self.current_ticket = {'Payment Received ': True}
                print(f"\nThank you for buying parking here Auto--> {self.license_plate_number}\nNow you have 15 minutes to leave or you will live here!!!!!!")
                self.parking_spaces.append(1)
                self.tickets.append(1)
                self.license_plate_numbers.remove(self.license_plate_number)
                
            else:
                self.current_ticket = {'Payment Received ': False}
                print('Please pay $10 or you are never leaving')



    def show_capacity(self):
        """Show's the capacity of the Garage"""
        current_capacity = len(self.parking_spaces)
    
        print(f"\nMax Capacity: 5 spaces\nCurrent Capacity: {len(self.parking_spaces)} spaces available\nDriver List: {self.license_plate_numbers}")   




driver3 = ParkingGarage('1il10o0o')
driver3.take_ticket()
driver5 = ParkingGarage('abc1234')
driver5.take_ticket()
driver5.show_capacity()
driver6 = ParkingGarage('ILVCT123')
driver6.take_ticket()
driver9 = ParkingGarage('jklp 85')
driver9.take_ticket()
driver4 = ParkingGarage('lk3444')
driver4.take_ticket()
driver4.show_capacity()
driver5.pay_for_parking(10)
driver5.show_capacity()



```

```python
    def run():
        while True:
            response = input("What would you like to do?\nEnter: 'c' to check capacity\n 'p' to park\n 'l' to leave\n 'q' to quit")

            if response.lower() == 'q':
                print('\nWhy did you even come here then?\nGOOOOOOOOOOOOOOOOOOD\nBYYYYYYYEEEEEEE!!!!!!')
                break

            elif response.lower() == 'p':
                license_plate_number = input("Enter your license plate number: ")
                driver = ParkingGarage(license_plate_number)
                driver.take_ticket()

            elif response.lower() == 'l':
                money_paid = int(input("Enter the amount you are paying ($10): "))
                driver.pay_for_parking(money_paid)
            
            elif response.lower() == 'c':
                driver.show_capacity()

    
run()
        
```
